```python
#Initialize
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Initialize

dfFormTop = pd.read_csv("../Darcy Analytics/Formation_Top.csv")
```

```python
dfFormTop.head()
```

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt


fig=plt.figure()
ax=plt.axes(projection='3d')
X = dfFormTop["X, m"]
Y = dfFormTop["Y, m"]
Z = dfFormTop["Top, m"]

S = ax.plot_trisurf(X, Y, -Z, cmap = 'viridis', edgecolor = 'none')
cbar = fig.colorbar(S)
cbar.ax.set_ylabel('Depth');
```

```python
fig, ax = plt.subplots(constrained_layout=True)
C = ax.tricontourf(X, Y, Z, levels = 30)
cbar = fig.colorbar(C)
cbar.ax.set_ylabel('Depth');
```
